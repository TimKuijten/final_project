```python
import os
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd


df_linkedin = []

# 1. Initialize the driver with options
options = webdriver.ChromeOptions()
options.add_argument("--incognito")
driver = webdriver.Chrome(options=options)

url = "https://www.linkedin.com/"
driver.get(url)

# 2. Load variables
my_email = os.getenv("email")
my_password = os.getenv("password")

# 3. Locate elements: sign-in
email = driver.find_element(By.ID, "session_key")
password = driver.find_element(By.ID, "session_password")

email.send_keys(my_email)
password.send_keys(my_password)

password.send_keys(Keys.RETURN)

time.sleep(10)

url = "https://www.linkedin.com/search/results/content/?heroEntityKey=urn%3Ali%3Aautocomplete%3A878310349&keywords=artificial%20intelligence&origin=SWITCH_SEARCH_VERTICAL&position=0&searchId=10ff84a6-6373-43c1-af08-6f7cb1218c42&sid=UQY"
driver.get(url)

# 5. Scrolling down
start = time.time()
wait = WebDriverWait(driver, 45)


try:
    while True:
        scroll_height = 10
        document_height_before = driver.execute_script("return document.documentElement.scrollHeight")
        driver.execute_script(f"window.scrollTo(0, {document_height_before + scroll_height});")
        time.sleep(8)
        document_height_after = driver.execute_script("return document.documentElement.scrollHeight") 
        if document_height_after == document_height_before:
            break
except:
    print('Couldnt scroll')

# 6. Getting elements
html = driver.execute_script("return document.body.outerHTML;")
soup = BeautifulSoup(html, "html.parser")

feed = soup.find_all("div", {"class":"search-results-container"})
posts = soup.find_all("div", {"class":"pt1 mb2 artdeco-card"})
rows = []
for post in posts:
    name = post.find("span", dir="ltr").text.strip()
    try:
        reactions = post.find("span", class_="social-details-social-counts__reactions-count").text.strip()
    except:
        reactions = '0'
    try:
        comments = post.find_all("button", class_="t-black--light social-details-social-counts__count-value t-12 hoverable-link-text")
    except:
        comments = '0'

    if comments:
        comments_text = comments[0].text.strip()
    else:
        comments_text = ""

    try:
        reposts = post.find_all("button", class_="ember-view t-black--light t-12 hoverable-link-text")
        if reposts:
            reposts_text = reposts[0].text.strip()
        else:
            reposts_text = ""
    except:
        reposts_text = ""

    try:
        description = post.find('div', {'class': 'feed-shared-update-v2__description-wrapper'})
        content = description.find('span').get_text(strip=True)
    except:
        content = ''

    try:
        comment_text = post.find('div', dir='ltr', class_='comments-comment-item__main-content feed-shared-main-content--comment t-14 t-black t-normal')
    except:
        comment_text = ''   

    try:
        comment_reaction = post.find_all('span', {'class': 'comments-comment-item__main-content'}).text.strip()   
    except:
        comment_reaction = 'NOT WORKING'  


    rows.append([name, reactions, comments_text, reposts_text, content])

df_linkedin = pd.DataFrame(rows, columns=['Name', 'Reactions', 'Comments', 'Reposts', 'Content'])

driver.quit()
```
